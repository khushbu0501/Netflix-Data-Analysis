## Handling Null  values

```python
netflix_df.isnull().sum()
```

 - In this dataset there is no null value and after glimpse on data we can recognize that they have substituted Not Given in place of null values 

```python
netflix_df.describe()
```

 - Only release_year column is of int type so data stats can be found of this column only . From min value it can be interpreted that First movie or show released in 1925 on netflix

```python
netflix1_df = pd.DataFrame(netflix_df)
```

```python
netflix1_df
```

```python
netflix1_df = netflix1_df.sort_values(by=['release_year'])
```

 - Here the netflix1_df is sorted on the basis of release year

```python
netflix1_df
```

```python
netflix1_df.date_added
```

 - Here the date on which the data is added in dataset, It is not in proper format of date. As we can't extract information of months and year seperately  from it . So we can convert it into proper format using to_datetime function of pandas library

```python
netflix1_df['date_added'] = pd.to_datetime(netflix1_df.date_added)
```

```python
netflix1_df['date_added']
```

```python
netflix1_df['year'] = pd.DatetimeIndex(netflix1_df.date_added).year
netflix1_df['month'] = pd.DatetimeIndex(netflix1_df.date_added).month
netflix1_df['day'] = pd.DatetimeIndex(netflix1_df.date_added).day
```

 - Here we can make three new columns of year,month and day from the date 

```python
netflix1_df
```

## Splitting the Dataset  

```python
type_counts = netflix1_df.type.value_counts()
type_counts
```

 - As the dataset's column type has only two values .So we can analyze it more easily if we split it on the basis of Netflix Movies and Netflix TV shows and create two new datasets

```python
netflix_tv = netflix1_df[netflix1_df.type=='TV Show']
```

# Exploratory analysis and Visualization

 - Here we analyse the data from different dimension and visualize it with some plots

 - Importing required packages for graph plotting

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

## Netflix Film Types:Movie Or TV Show

```python
type_counts = netflix1_df.type.value_counts()
```

```python
type_counts
```

 - Plotting the graph on the count of TV show and Movie 

```python
plt.figure(figsize=(7,5))
sns.countplot(netflix1_df.type,palette="pastel");
plt.title("Count of Movies and TV Shows")
plt.xlabel("Type(Movie/TV Show)")
plt.ylabel("Total Count")
plt.show()
```

 - The above countplot shows that count of Movies on netflix is more than TV shows and same information is conveyed by below pie chart and percentage is also indicated

```python
plt.figure(figsize=(12,6))
plt.title('Percentage of materials watched')
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', startangle=180);
```

## Variations in terms of release Year

```python
plt.title("Distribution in terms of release_year")
plt.hist(netflix1_df.release_year,bins=10);
```

## Content Added Each Year

```python
netflix_year = netflix1_df['year'].value_counts().to_frame().reset_index().rename(columns={'index':'years','year':'count'})
netflix_year
```

```python
netflix_movie_year = netflix_movie_df['year'].value_counts().to_frame().reset_index().rename(columns={'index':'years','year':'count'})
netflix_movie_year
```

```python
netflix_tv_year = netflix_tv['year'].value_counts().to_frame().reset_index().rename(columns={'index':'years','year':'count'})
netflix_tv_year
```

 - Line plot showing the variation of content added on each year on netflix. 

```python
fig, ax = plt.subplots(figsize=(10, 6))
sns.lineplot(data=netflix_year, x='years', y='count')
sns.lineplot(data=netflix_movie_year, x='years', y='count')
sns.lineplot(data=netflix_tv_year, x='years', y='count')
ax.set_xticks(np.arange(2008, 2020, 1))
plt.title("Total content added each year ")
plt.legend(['Total','Movie','TV Show'])
plt.ylabel("Releases")
plt.xlabel("Year")
plt.show()
```

```python
plt.title('year and Month distribution when added')
sns.scatterplot(x=netflix1_df.year,y= netflix1_df.month,hue=netflix_df.type,s=100);
```

## Popular Tags 

```python
filtered_tags = netflix1_df.set_index('title').listed_in.str.split(', ', expand=True).stack().reset_index(level=1, drop=True);

plt.figure(figsize=(7,9))
g = sns.countplot(y = filtered_tags, order=filtered_tags.value_counts().index[:20])
plt.title('Top 20 Tags on Netflix')
plt.xlabel('Titles')
plt.ylabel('Tags')
plt.show()
```

 - Most of the netflix movies and TV Shows has listed in International Movies
